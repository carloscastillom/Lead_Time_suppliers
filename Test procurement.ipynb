```python
import pandas as pd
path= r'H:\Jupyter_test\supplierLeadTime\SLT.csv'
```

```python
#encodin utf16
#use the first column as colum names
#parse the following columns as date
test = pd.read_csv(path, encoding='utf16', on_bad_lines='warn', index_col=False)  
test['PO Date'] = pd.to_datetime(test['PO Date'],
                                          format="%d.%m.%Y %H:%M")
```

```python
#keep all notnull for the goods receipt posting date - only analysis over what have been calculated 
test = test[test['Q_r'].notna()]
#keep all null for the Goods_Returned - make sure we have not returned
test = test[test['Goods_Returned'].isna()]
```

```python
test = test.drop(['#','Document Internal ID','Estimated Delivery Date','Goods Receipt Internal ID','Goods_Returned','Unnamed: 13'], axis=1)
test.info()
```

```python
test['Val_Diff'] = test['Goods Receipt Posting Date'] - test['PO Date']
test.head(20)
```
